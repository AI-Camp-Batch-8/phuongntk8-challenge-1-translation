```python
import pandas as pd
```

```python
df = pd.read_excel('menu.xlsx')
```

```python
df
```

```python
df_vi_en = df.iloc[:, 1:3]
df_vi_en
```

```python
df_vi_en_shuffled = df_vi_en.sample(frac=1,random_state=200) 
df_vi_en_shuffled
```

```python
len(df_vi_en_shuffled)
```

```python
num_samples = len(df_vi_en_shuffled)
df_train = df_vi_en_shuffled[:int(num_samples*0.8)]
df_test = df_vi_en_shuffled[int(num_samples*0.8):int(num_samples*0.9)]
df_dev = df_vi_en_shuffled[int(num_samples*0.9):]
```

```python
df_train
```

```python
df_test
```

```python
df_dev
```

```python
with open('./train.vi', 'w') as f:
    for i in range(len(df_train)):
        if str(df_train.iloc[i, 0]) == '':
            f.write('NaN\n')
        else:
            line = str(df_train.iloc[i, 0])
            line = line.strip().split()
            line = ' '.join(line)
            f.write(line + '\n')
```

```python
with open('./train.en', 'w') as f:
    for i in range(len(df_train)):
        if str(df_train.iloc[i, 1]) == '':
            f.write('NaN\n')
        else:
            line = str(df_train.iloc[i, 1])
            line = line.strip().split()
            line = ' '.join(line)
            f.write(line + '\n')
```

```python
with open('./test.vi', 'w') as f:
    for i in range(len(df_test)):
        if str(df_test.iloc[i, 0]) == '':
            f.write('NaN\n')
        else:
            line = str(df_test.iloc[i, 0])
            line = line.strip().split()
            line = ' '.join(line)
            f.write(line + '\n')


with open('./test.en', 'w') as f:
    for i in range(len(df_test)):
        if str(df_test.iloc[i, 1]) == '':
            f.write('NaN\n')
        else:
            line = str(df_test.iloc[i, 1])
            line = line.strip().split()
            line = ' '.join(line)
            f.write(line + '\n')
```
